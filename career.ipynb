```python
#Converting to integer type.
df['experience_years']=df['experience_years'].str.replace("years","").astype('Int64')
df['experience_years']
```

```python
#Dropping useless columns
df.drop(columns=['user_id','name'], inplace=True)
```

```python
# Map categorical risk tolerance levels to numerical values
df['risk_tolerance']=df['risk_tolerance'].map({'low':0, 'medium':1, 'high':2})
df['risk_tolerance'].dtype
```

```python
# Define numeric and categorical features for preprocessing
numeric_features = [
    'age', 'experience_years', 'openness', 'conscientiousness', 'extraversion',
    'agreeableness', 'neuroticism', 'technical_skill', 'data_reasoning_skill',
    'communication_skill', 'problem_solving_skill', 'leadership_skill',
    'creative_thinking_skill', 'interest_technology', 'interest_business',
    'interest_creative', 'interest_health_social', 'interest_research_academic',
    'risk_tolerance'
]

categorical_features = [
    'education_level', 'field_of_study', 
    'preferred_work_style', 'preferred_environment'
]
df_for_stats = df.copy() 
```

```python
#Checking for missing values
df.isna().sum()
```

```python
#Printing unique values in features.
l = ['age','education_level','field_of_study', 'experience_years', 'openness', 'conscientiousness', 'extraversion', 'agreeableness', 'neuroticism', 'technical_skill', 'data_reasoning_skill', 'communication_skill', 'problem_solving_skill', 'leadership_skill', 'creative_thinking_skill', 'interest_technology', 'interest_business', 'interest_creative', 'interest_health_social', 'interest_research_academic', 'preferred_work_style', 'preferred_environment', 'risk_tolerance']
for i in l:
    print(i, ":", df[i].unique(), "\n")
```

```python
#Filling missing values
median_cols = [
    'age', 'experience_years', 'openness', 'conscientiousness',
    'extraversion', 'agreeableness', 'neuroticism',
    'technical_skill', 'data_reasoning_skill', 'communication_skill',
    'problem_solving_skill', 'leadership_skill', 'creative_thinking_skill',
    'interest_technology', 'interest_business', 'interest_creative',
    'interest_health_social', 'interest_research_academic'
]

for col in median_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
mode_cols = [
    'education_level', 'field_of_study',
    'preferred_work_style', 'preferred_environment',
    'risk_tolerance'
]

for col in mode_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
#IQR
num_cols = df.select_dtypes(include=['int64','float64']).columns
outlier_len = {}
for i in num_cols:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1 
    outliers = df[(df[i] < Q1 - 1.5*IQR) | (df[i] > Q3 + 1.5*IQR)]
    outlier_len[i] = len(outliers)
outlier_len
```

```python
from scipy.stats import chi2_contingency
target = "career_cluster"
categorical_col=df.select_dtypes(include=['object']).columns
results = []
for col in categorical_col:
    contingency_table = pd.crosstab(df[col], df[target])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    results.append({
        "feature": col,
        "chi_square": round(chi2, 3),
        "p_value": round(p, 6),
        "degrees_of_freedom": dof,
        "significant_at_0.05": p < 0.05
    })
chi_square_df = pd.DataFrame(results).sort_values("p_value")
chi_square_df
```

```python
#Feature Engineering
interest_cols = ['interest_technology', 'interest_business', 'interest_creative','interest_health_social', 'interest_research_academic']
df['max_interest'] = df[interest_cols].max(axis=1)
df['tech_vs_business'] = df['interest_technology'] - df['interest_business']
df['creative_vs_social']= df['interest_creative'] - df['interest_health_social']
df['tech_skill_vs_comm']= df['technical_skill'] - df['communication_skill']
df['openness_creative']= df['openness']   * df['creative_thinking_skill']
df['extraversion_lead']= df['extraversion'] * df['leadership_skill']
df['is_stem']= df['field_of_study'].str.contains('computer|engineering|it|math|physics|chemistry|biology|data', case=False, na=False).astype(int)
```

```python
#Encoding
categorical_col=df.select_dtypes(include=['object']).columns
encode_col=categorical_col.drop('career_cluster')
df=pd.get_dummies(df,columns=encode_col,drop_first=True)
```

```python
#Imbalance check
counts = df['career_cluster'].value_counts()
percent = df['career_cluster'].value_counts(normalize=True) * 100
imbalance_df = pd.DataFrame({
    'count': counts,
    'percentage': percent.round(2)
})
imbalance_df
```

```python
imbalance_ratio = counts.max() / counts.min()
imbalance_ratio
```

```python
x=df.drop(columns='career_cluster')
y=df['career_cluster']
```

```python
#Train-test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,stratify=y,random_state=42)
x_train
```

```python
#Model training
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=600, max_depth=11, max_features=0.35, min_samples_leaf=3, ccp_alpha=0.0001, random_state=42, n_jobs=-1)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
#Model evaluation
from sklearn.metrics import classification_report,accuracy_score
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
# Get predictions for both sets
train_preds = model.predict(x_train)
test_preds = model.predict(x_test)

# Calculate accuracy
from sklearn.metrics import accuracy_score
train_acc = accuracy_score(y_train, train_preds)
test_acc = accuracy_score(y_test, test_preds)
print(f"Training Accuracy: {train_acc * 100:.2f}%")
print(f"Testing Accuracy: {test_acc * 100:.2f}%")
```

```python
#Report
report=classification_report(y_test,y_pred)
print(report)
```

```python
importances = pd.Series(model.feature_importances_,index=x_train.columns)
top_features = importances.nlargest(15)
plt.figure(figsize=(10, 6))
top_features.sort_values().plot(kind='barh')
plt.title("Top 15 Features")
plt.xlabel("Importance Score")
plt.show()
```

```python
#Front end
import joblib
joblib.dump(model, 'career_model.joblib') # Save the trained model
feature_columns = list(x.columns)
joblib.dump(feature_columns, 'feature_columns.joblib') # Save feature columns
numeric_medians = df_for_stats[numeric_features].median().to_dict()
joblib.dump(numeric_medians, 'numeric_medians.joblib')# Save numeric medians
df_original = pd.read_csv('career.csv')
cat_modes = {col: df_original[col].mode()[0] for col in categorical_features}
joblib.dump(cat_modes, 'cat_modes.joblib')# Save categorical modes
print("\nFiles created:")
print("  - career_model.joblib")
print("  - feature_columns.joblib")
print("  - numeric_medians.joblib")
print("  - cat_modes.joblib")
```
